```python
import reverse_geocoder

import pandas



def reverse(coordinates):
    res = reverse_geocoder.search(coordinates)
    return (res[0]['name'])
    
data = pandas.read_csv(r'C:\Users\ooagboola\OneDrive - Union Bank\Documents\Location_data.csv')
data
```

```python
data['geom'] = data.apply(lambda row: (str(row.Latitude),str(row.Longitude)),axis=1)
data
```

```python
data['Address'] = data.apply(lambda row: reverse(row.geom), axis=1)
data
```
